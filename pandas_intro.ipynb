##### Mean 
Mathematical expectation or average, is the central value of a discrete set of numbers: specifically, the sum of the values divided by the number of values. 

##### STD (Standart Deviation)
SD is a measure of the amount of variation or dispersion of a set of values. A low standard deviation indicates that the values tend to be close to the mean (also called the expected value) of the set, while a high standard deviation indicates that the values are spread out over a wider range.

#### Histogram
A histogram is an accurate representation of the distribution of numerical data. It is an estimate of the probability distribution of a continuous variable.

```python
#histogram of the dataset
csv['SalePrice'].hist(bins=20)
# less bin size better
```

```python
#skewness and kurtosis
print("Skewness: %f" % csv['SalePrice'].skew())
print("Kurtosis: %f" % csv['SalePrice'].kurt())
```

#### Skewness
Skewness is a measure of symmetry, or more precisely, the lack of symmetry. A distribution, or data set, is symmetric if it looks the same to the left and right of the center point.
#### Kurtosis
Kurtosis is a measure of whether the data are heavy-tailed or light-tailed relative to a normal distribution. That is, data sets with high kurtosis tend to have heavy tails, or outliers. Data sets with low kurtosis tend to have light tails, or lack of outliers.

```python
#scatter plot grlivarea/saleprice
#we use  scatter plot only for continuos data, ex. area, price and etc.
#in our case we get scatter plot for area of the house and the sale price of the house.
var = 'GrLivArea'
data = pandas.concat([csv['SalePrice'], csv[var]], axis=1)
data.plot.scatter(x=var, y='SalePrice', ylim=(0,500000));
```

```python
#example for scatter plot with number of bedrooms and saleprice
#cause it's the categorical data, scatter plot doesn't give any info.
var = 'BedroomAbvGr'
data = pandas.concat([csv['SalePrice'], csv[var]], axis=1)
data.plot.scatter(x=var, y='SalePrice', ylim=(0,500000));
```
